```python
data['fecha'] = pd.to_datetime({'year':data.date, 'month': data.mes, 'day': data.dia}) 
data
```

```python
data = data.drop(columns= ["date", "dia", "mes"])
data = data.rename(columns= {"fecha": "ds", "Life_expectancy_at_birth_total": "y"})
data
```

```python
#Reemplazamos el código de cada serie por su nombre
data = data.rename(columns={"countryiso3code": "Country_code", "SP.ADO.TFRT": "Adolescent_fertility", "SP.DYN.CBRT.IN": "Birth_rate", "SH.XPD.CHEX.GD.ZS": "Health_expenditure", "SH.XPD.CHEX.PC.CD": "Health_expend_per_capita", "SP.DYN.TFRT.IN": "Fertility_rate_total", "SP.DYN.LE00.IN": "Life_expectancy_at_birth_total", "SP.DYN.IMRT.IN": "Infant_Mortality_rate", "SP.POP.GROW": "Population_growth", "SP.POP.TOTL": "Total_Population", "SH.STA.SUIC.P5": "Suicide_rate", "NY.GDP.PCAP.CD": "GDP_per_capita", "NY.GDP.MKTP.CD": "GDP", "ncd_bmi_25a": "Adults_Overweight", "ncd_hyp_prevalence_a": "Adults_Hypertension"})
data.head()
```

```python
from prophet import Prophet
```

```python
# Prophet trabaja la columna fecha con el nombre 'ds' y la columna con las etiquetas con el nombre 'y'
data = data.drop(columns= ["Country_code", "country_value", "Adolescent_fertility", "Birth_rate", "Health_expenditure", "Health_expend_per_capita", "Fertility_rate_total", "Infant_Mortality_rate", "Population_growth", "Total_Population", "Suicide_rate", "GDP_per_capita", "GDP", "Adults_Overweight", "Adults_Hypertension"])
data = data.rename(columns= {"date": "ds", "Life_expectancy_at_birth_total": "y"})
data
```

```python
# data = data.sort_values(by="ds", ascending=False)
# data
```

```python
# Instanciamos un modelo de Prophet

m = Prophet(yearly_seasonality= True, weekly_seasonality=False, daily_seasonality=False, seasonality_mode="multiplicative")  

# Entrenamos el modelo
m.fit(data)

# Construimos un dataframe con los x valores futuros que querámos predecir. En este caso, 4 años.
future = m.make_future_dataframe(periods=365) 
future.tail()
```

```python
# Predecimos
forecast = m.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
fig1 = m.plot(forecast)
```

```python
# plot forecasts

prophet.plot(forecast, xlabel='Año', ylabel='Esperanza de Vida', figsize = (16,6))
plt.title('Predicción de Esperanza de Vida para los próximos años de nuestro dataset', fontsize = 18)
plt.show()
```

```python
fig2 = m.plot_components(forecast)
```

```python
##Evaluamos el modelo
from prophet.diagnostics import cross_validation
from prophet.diagnostics import performance_metrics
from prophet.plot import plot_cross_validation_metric
```

```python
crossval = cross_validation(prophet, horizon = '365 days')
```

```python
crossval
```

```python
df_p = performance_metrics(crossval)
df_p.describe()
```

```python
fig = plot_cross_validation_metric(crossval, metric='rmse')
```

```python
print('El RMSE es:', df_p.rmse.mean())
```

```python
df_p = performance_metrics(crossval, rolling_window = 1)
```
