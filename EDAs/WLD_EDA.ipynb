# World -Exploratory Data Analysis 

This region contains countries with high population around the world that didn't make it into the regions we are studying or simply did not belong to them. but for analysis purposes we decided to include them in this region WLD - World those countries are:

- AUS - Australia

- CHN - China

- GBR - United Kingdom

- IDN - Indonesia

- RUS - Russia

- USA - United States

```
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Main DataFrame

```
main_df = pd.read_csv('/work/Datasets/WLD/WLD.csv')
main_df = main_df.rename(columns={
    'countryiso3code' : 'country', 
    'SP.ADO.TFRT':'adolescent_fertility_rate', 
    'SP.DYN.CBRT.IN':'birth_rate',
    'SP.DYN.LE00.IN':'life_expectancy',
    'SH.XPD.CHEX.PC.CD':'health_expenditure_per_capita'
})
main_df.head()
```

```
#Table information with the columns and their data types
size = main_df.shape
print(f'table dimensions, rows: {size[0]} columns: {size[1]}')

main_df.info()
```

## Adolescent Fertility Rate - Region: WLD

```
ad_df = main_df[['date', 'country', 'adolescent_fertility_rate']]
ad_df.head()
```

```
#Empty Values
empty_values = ad_df.adolescent_fertility_rate.isna().sum()
print(f'The empty values total is: {empty_values}')
ad_df[ad_df.adolescent_fertility_rate.isna()]
```

```
#Adolescent fertility through time plot
for country in ad_df.country.unique():
    plt.plot(ad_df.date[ad_df.country == country], ad_df.adolescent_fertility_rate[ad_df.country == country], label=country)
plt.legend()
plt.show()
```

## Birth Rate - Region: WLD

```
br_df = main_df[['date', 'country', 'birth_rate']]
br_df.head()
```

```
# Empty birth_rate column values
empty_values = br_df.birth_rate.isna().sum()
print(f'The empty values total is: {empty_values}')
br_df[br_df.birth_rate.isna()]
```

```
#Birth rate through time plot
for country in ad_df.country.unique():
    plt.plot(br_df.date[br_df.country == country], br_df.birth_rate[br_df.country == country], label=country)
plt.legend()
plt.show()
```

## Life Expectancy - Region: WLD

```
le_df = main_df[['date', 'country', 'life_expectancy']]
le_df.head()
```

```
#Missing values
empty_values = le_df.life_expectancy.isna().sum()
print(f'The empty values total is: {empty_values}')
le_df[le_df.life_expectancy.isna()]
```

```
# Life expectancy over time plot
for country in ad_df.country.unique():
    plt.plot(le_df.date[le_df.country == country], le_df.life_expectancy[le_df.country == country], label=country)
plt.legend()
plt.show()
```

## Health Expenditure per Capita- Region: WLD

```
he_df = main_df[['date', 'country', 'health_expenditure_per_capita']]
he_df.head(15)
```

```
empty_values = he_df.health_expenditure_per_capita.isna().sum()
print(f'The missing values total is: {empty_values}')
print(he_df.health_expenditure_per_capita.isna().groupby(he_df['country']).sum())

he_df[he_df.health_expenditure_per_capita.isna()]
```

```
# Health expenditure over time
for country in he_df.country.unique():
    plt.plot(he_df.date[he_df.country == country], he_df.health_expenditure_per_capita[he_df.country == country], label=country)
plt.legend()
plt.show()
```
