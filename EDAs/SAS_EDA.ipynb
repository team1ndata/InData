# South Asia Region - EDA

```
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Main DataFrame

```
main_df = pd.read_csv('/work/Datasets/SAS/SAS.csv')
main_df = main_df.rename(columns={
    'countryiso3code' : 'country', 
    'SP.ADO.TFRT':'adolescent_fertility_rate', 
    'SP.DYN.CBRT.IN':'birth_rate',
    'SP.DYN.LE00.IN':'life_expectancy',
    'SH.XPD.CHEX.PC.CD':'health_expenditure_per_capita'
})
main_df.head()
```

```
#General overview of the table
size = main_df.shape
print(f'Rows: {size[0]} Columns: {size[1]}')
main_df.info()
```

## Adolescent Fertility Rate - Region: SAS

```
af_df = main_df[['date', 'country', 'adolescent_fertility_rate']]
af_df.head()
```

```
#Empty values
empty_values = af_df.adolescent_fertility_rate.isna().sum()
print(f'The column "adolescent_fertility_rate" presents {empty_values} empty values')
af_df[af_df.adolescent_fertility_rate.isna()]
```

```
# Adolescent fertility though time plot
for country in af_df.country.unique():
    plt.plot(af_df.date[af_df.country == country], af_df.adolescent_fertility_rate[af_df.country == country], label=country)
plt.legend()
plt.show()
```

## Birth Rate - Region: SAS

```
#Filter main dataframe 
br_df = main_df[['date', 'country', 'birth_rate']]
br_df.head()
```

```
#Find missing values
empty_values = br_df.birth_rate.isna().sum()
print(f'The total empty values in the column "birth_rate" is: {empty_values}')
br_df[br_df.birth_rate.isna()]
```

```
#Birth rate Through time 
for country in af_df.country.unique():
    plt.plot(br_df.date[br_df.country == country], br_df.birth_rate[br_df.country == country], label=country)
plt.legend()
plt.show()
```

## Life Expectancy - Region: SAS

```
#Filter main dataframe
le_df = main_df[['date', 'country', 'life_expectancy']]
le_df.head()
```

```
#Missing values
empty_values = le_df.life_expectancy.isna().sum()
print(f'The total empty values in the column "life_expectancy" is: {empty_values}')
le_df[le_df.life_expectancy.isna()]
```

```
#Life expectancy Through time 
for country in af_df.country.unique():
    plt.plot(le_df.date[le_df.country == country], le_df.life_expectancy[le_df.country == country], label=country)
plt.legend()
plt.show()
```

## Health Expenditure - Region: SAS

```
he_df = main_df[['date', 'country', 'health_expenditure_per_capita']]
he_df.head()
```

```
#Empty values
empty_values = he_df.health_expenditure_per_capita.isna().sum()
print(f'The missing values total is: {empty_values}')
print(he_df.health_expenditure_per_capita.isna().groupby(he_df['country']).sum())

he_df[he_df.health_expenditure_per_capita.isna()]

# AFG present more missing values compared to other countries in the region
```

```
#Life expectancy Through time 
for country in af_df.country.unique():
    plt.plot(he_df.date[he_df.country == country], he_df.health_expenditure_per_capita[he_df.country == country], label=country)
plt.legend()
plt.show()
```
