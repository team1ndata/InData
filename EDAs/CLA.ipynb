# Latin America & the Caribbean EDA

### Import libraries

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

#### Main DataFrame


```
main_df = pd.read_csv('/work/Datasets/CLA/CLA.csv')
main_df = main_df.rename(columns={
    'countryiso3code' : 'country', 
    'SP.ADO.TFRT':'adolescent_fertility_rate', 
    'SP.DYN.CBRT.IN':'birth_rate',
    'SP.DYN.LE00.IN':'life_expectancy',
    'SH.XPD.CHEX.PC.CD':'health_expenditure_per_capita'
})
main_df.head()
```

```
#Information of all the columns and their data types
main_df.info()
```

## Adolescent fertility rate - Region: CLA

```
adolescent_df = main_df[['date', 'country', 'adolescent_fertility_rate']]
adolescent_df.head()
```

```
# Empty values in the dataframe
number_of_empty_cells = adolescent_df.adolescent_fertility_rate.isna().sum()
print(f'The number of empty cells in the fertility column is: {number_of_empty_cells}')
adolescent_df[adolescent_df.adolescent_fertility_rate.isna()]
```

```
for country in adolescent_df.country.unique():
    plt.plot(adolescent_df.date[adolescent_df.country == country], adolescent_df.adolescent_fertility_rate[adolescent_df.country == country], label=country)
plt.legend()
plt.show()
```

## Birth rate - Region: CLA

```
birth_rate_df = main_df[['date', 'country', 'birth_rate']]
birth_rate_df.head()
```

```
# Empty values in birth_rate column
empty_values = birth_rate_df.birth_rate.isna().sum()
print(f'Empty cells: {empty_values}')
birth_rate_df[birth_rate_df.birth_rate.isna()]
```

```
#Graphing the birth rate across the time
for country in adolescent_df.country.unique():
    plt.plot(birth_rate_df.date[birth_rate_df.country == country], birth_rate_df.birth_rate[birth_rate_df.country == country], label=country)
plt.legend()
plt.show()
```

## Life Expectancy - Region: CLA

```
le_df = main_df[['date', 'country', 'life_expectancy']]
le_df.head()
```

```
# Check for empty values in life expectancy column
empty_values = le_df.life_expectancy.isna().sum()
print(f'Empty cells: {empty_values}')
le_df[le_df.life_expectancy.isna()]
```

```
#Life expectancy through the years plot
for country in main_df.country.unique():
    plt.plot(le_df.date[le_df.country == country], le_df.life_expectancy[le_df.country == country], label=country)

plt.legend()
plt.show()
```
